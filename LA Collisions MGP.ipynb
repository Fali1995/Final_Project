```python
my_data = my_data[['Date Occurred', 'Time Occurred','Location', 'Area Name']]
```

```python
my_data['Date Occurred'] = pd.to_datetime(my_data['Date Occurred'])
```

```python
my_data.head()
```

```python
my_data.dtypes
```

```python
my_data = my_data.set_index('Date Occurred')
my_data.head()
```

```python
my_data['Year'] = my_data.index.year
my_data['Month'] = my_data.index.month
my_data['Weekday Name'] = my_data.index.weekday_name
# Display a random sampling of 5 rows
my_data.sample(5, random_state=0)
```

```python
my_data['Location'] = [ast.literal_eval(d) for d in my_data.Location]
my_data['longitude'] = [d['longitude'] for d in my_data.Location]
my_data['latitude'] = [d['latitude'] for d in my_data.Location]
my_data.longitude = my_data.longitude.astype(float)
my_data.latitude = my_data.latitude.astype(float)
```

```python
my_data.head()
```

```python
my_data_2019 = my_data[my_data.Year==2019]

print("2019 monthly average: {}".format(my_data_2019.resample('M').size().mean()))
```

```python
plt.figure(figsize=(15,20))
sns.scatterplot(x='longitude', y='latitude', hue='Weekday Name', palette=p, data=my_data_2019.sample(3000))
plt.title("Average month in 2019")
```
